```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


from google.colab import files
uploaded = files.upload()


df = pd.read_csv("DataAnalyst.csv")


df.head()


print("Size:", df.shape)


print(":\n", df.isnull().sum())


df = df.drop_duplicates()


df.columns = [col.lower() for col in df.columns]


print(df.describe())

df.hist(bins=30, figsize=(15, 10))
plt.show()

plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.show()

summary = """
### Summary & Insights

1. The dataset was successfully loaded and inspected.
2. Missing values were checked and can be handled if necessary.
3. Duplicate rows were removed to improve data quality.
4. The distributions of numerical values show some irregularities and possible
outliers.
5. The correlation analysis indicates relationships between some variables.
6. Before applying any modeling, further preprocessing such as handling
missing data, encoding categorical values, and scaling may be required.
"""

print(summary)


```
